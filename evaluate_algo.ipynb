```python
import cv2
import json
import pandas as pd

from utils import full_ocr, evaluate_image_ocr, full_ocr_v2
```

```python
pd.options.display.max_rows = 999
```

```python
# Images with end results to evaluate E2E algo
with open('uids_items.json', 'r') as f:
    uids_items = json.load(f)
uids_with_items = [item['uid'] for item in uids_items]
print('Nb images with end res:', len(uids_with_items))
```

```python
data = []
for d in uids_items:
    uid = d['uid']

    print(uid)

    with open('texts_coordinates/' + uid + '.json', 'r') as f:
        texts_coordinates = json.load(f)

    image = cv2.imread(f'cuentas/{uid}.jpeg')
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    try:
        items_found = full_ocr(image, texts_coordinates)
    except:
        items_found = []

    print(items_found)

    score = evaluate_image_ocr(d['items'], items_found)
    print(score)

    data.append({
        'uid': uid,
        'items_found': items_found,
        'score': score
    })
```

```python
scores = pd.DataFrame(data)
```

```python
new_data = []
for d in uids_items:
    uid = d['uid']

    print(uid)

    with open('texts_coordinates/' + uid + '.json', 'r') as f:
        texts_coordinates = json.load(f)

    image = cv2.imread(f'cuentas/{uid}.jpeg')
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    try:
        items_found, std_angles = full_ocr_v2(image, texts_coordinates)
    except:
        items_found = []

    print(items_found)

    score = evaluate_image_ocr(d['items'], items_found)
    print(score)

    new_data.append({
        'uid': uid,
        'items_found_2': items_found,
        'score_2': score,
        'std_angles': std_angles
    })
```

```python
agg_scores = scores.merge(pd.DataFrame(new_data), on='uid')
```

```python
agg_scores[agg_scores['score'] > agg_scores['score_2']]
```
